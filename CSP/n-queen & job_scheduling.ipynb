### N-QUEEN PROBLEM

```python
problem = Problem()
n = int(input("Enter N: "))
rows = range(n)
cols = range(n)
problem.addVariables(cols, rows)
for i in cols:
    for j in cols:
        if i < j:
            problem.addConstraint(lambda row1, row2, col1=i, col2=j:
                                  abs(row1-row2) != abs(col1-col2) and row1 != row2, (i,j)) #checking for diagonal,rows and column

solutions = problem.getSolutions()
length = len(solutions)
print("No of Solutions: ", length)
for sol in solutions:
    print("Solution: ")
    arr = [["_" for i in range(n)] for j in range(n)]
    for i, (j,k) in enumerate(sol.items()):
        arr[j][k] = 'X'
    for i in range(n):
        for j in range(n):
            print(arr[i][j],end=" ")
        print()
```

## Job scheduling

### Assign 5 workers W1,W2,W3,W4,W5 to 5 products such that each worker works on one product and each product is produced by one worker. Effectiveness of production is given by the following table (e.g., worker W1 produces P1 with effectivness 7) and the total effectiveness must be 'n' atleast
####          p1  p2  p3  p4  p5
####   w1   7    1    3    4    2
####   w2   8    2    5    1    3
####   w3   4    3    7    2    1
####   w4   3    1    4    6    2
####   w5   1    3    2    4    5

```python
problem2 = Problem()
problem2.addVariables('a',list(enumerate([7,1,3,4,2])))
problem2.addVariables('b',list(enumerate([8,2,5,1,3])))
problem2.addVariables('c',list(enumerate([4,3,7,2,1])))
problem2.addVariables('d',list(enumerate([3,1,4,6,2])))
problem2.addVariables('e',list(enumerate([1,3,2,4,5])))
n = int(input("Enter effectiveness: ")) #25
def sum_constraint(w1,w2,w3,w4,w5):
    if w1[1]+w2[1]+w3[1]+w4[1]+w5[1] >= n:
        return True
def diff_constraint(w1,w2,w3,w4,w5):
    if w1[0]!= w2[0] and w1[0]!= w3[0] and w1[0]!= w4[0] and w1[0]!= w5[0] and w2[0]!= w3[0] and w2[0]!= w4[0] and w2[0]!= w5[0] and w3[0]!= w4[0] and w3[0]!= w5[0] and w4[0]!= w5[0]:
        return True

problem2.addConstraint(sum_constraint,['a','b','c','d','e'])
problem2.addConstraint(diff_constraint,['a','b','c','d','e'])
solutions = problem2.getSolutions()
length = len(solutions)
print("No of Solutions: ", length)
for sol in solutions:
    for i, (j,k) in enumerate(sol.items()):
        print(j,":",k[1], end=", ")
    print("\n")
```
